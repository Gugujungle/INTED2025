## -  Crie um algoritmo que receba a idade do usuário, e depois verifique em qual faixa etária está à idade:

até 12 - Você é uma Criança!

de 13 á 18 - Você é um Adolescente!

de 19 à 60 - Você é um Adulto!

acima de 60 - Voce é um Idoso!

```python
idade = int(input("Digite sua idade: "))

if idade <= 12:
  print("Você é uma Criança!")
elif idade >= 13 and idade <= 18:
  print("Você é um Adolescente!")
elif idade >= 19 and idade <= 60:
  print("Você é um Adulto!")
else:
  print("Voce é um Idoso!")
```

##Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
else:
  horas_normais = 40
  horas_extras = horas_trabalhadas - 40
  salario = (horas_normais * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

print(f"O salário total é: R${salario:.2f}")
```

##Calculadora V25

```python
opcao = -1

while opcao != 0:
    print("\n===== CALCULADORA V25 =====")
    print("1 - Múltiplos de 5")
    print("2 - Números pares")
    print("3 - Números primos")
    print("4 - Conversor de temperatura")
    print("0 - Sair")
    print("----------------------------")

    opcao = int(input("Digite a opção: "))

    # ------------------------------
    # MÚLTIPLOS DE 5
    # ------------------------------
    if opcao == 1:
        n = int(input("Digite o número limite: "))
        print(f"Múltiplos de 5 até {n}:")

        i = 0
        while i <= n:
            if i % 5 == 0:
                print(i)
            i += 1

    # ------------------------------
    # NÚMEROS PARES
    # ------------------------------
    elif opcao == 2:
        n = int(input("Digite o número limite: "))
        print(f"Números pares até {n}:")

        i = 0
        while i <= n:
            if i % 2 == 0:
                print(i)
            i += 1

    # ------------------------------
    # NÚMEROS PRIMOS
    # ------------------------------
    elif opcao == 3:
        n = int(input("Digite o número limite: "))
        print(f"Números primos até {n}:")

        i = 2
        while i <= n:
            primo = True
            j = 2

            while j < i:
                if i % j == 0:
                    primo = False
                    break
                j += 1

            if primo:
                print(i)

            i += 1

    # ------------------------------
    # CONVERSOR DE TEMPERATURA
    # ------------------------------
    elif opcao == 4:
        print("1 - Celsius para Fahrenheit")
        print("2 - Fahrenheit para Celsius")

        op = int(input("Escolha: "))

        if op == 1:
            c = float(input("Digite a temperatura em Celsius: "))
            f = (c * 9/5) + 32
            print(f"{c}°C = {f:.2f}°F")

        elif op == 2:
            f = float(input("Digite a temperatura em Fahrenheit: "))
            c = (f - 32) * 5/9
            print(f"{f}°F = {c:.2f}°C")

        else:
            print("Opção inválida!")

    # ------------------------------
    # SAIR
    # ------------------------------
    elif opcao == 0:
        print("Saindo...")

    else:
        print("Opção inválida!")
```
